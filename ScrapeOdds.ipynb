```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import datetime

import pandas as pd
```

```python
year = 2018

thisdate = datetime.date(year, 1,1)


resultsdf = pd.DataFrame(columns = ['date', 'away', 'away_sp', 'away_score', 'home', 'home_sp', 'home_score', 'away_opener_odds', 'home_opener_odds'])
```

```python
driver = webdriver.Chrome()
```

```python
while thisdate.year == year:

    print(thisdate.isoformat())

    driver.get("https://www.sportsbookreview.com/betting-odds/mlb-baseball/?date=" + thisdate.isoformat())
    games = driver.find_elements(By.XPATH, '//div[@id="tbody-mlb"]/*')

    for game in games:
        gameid = game.find_elements(By.XPATH, '*[contains(@class, "BettingOptionContainer")]/*')
        odds = game.find_element(By.XPATH, '*[contains(@class, "containerTable")]')
        
        
        descriptors = gameid[1].text.splitlines()
        descriptors = [a for a in descriptors if '-' not in a]
        if len(descriptors) == 6:
            descriptors = ','.join(descriptors)
        else:
            #figure out what side is missing
            if '(L)' not in ','.join(descriptors[0:3]) and '(R)' not in ','.join(descriptors[0:3]):
                descriptors.insert(1, '')
            if '(L)' not in ','.join(descriptors[-3:]) and '(R)' not in ','.join(descriptors[-3:]):
                descriptors.insert(-1, '')

            descriptors = ','.join(descriptors)
            

        consensus = odds.find_elements(By.XPATH, 'div/div/*[contains(@class, "consensusColumn")]')[1]
        consensusopeners = consensus.text.splitlines()
        consensusopeners = ','.join(consensusopeners)

        stringline = ','.join([thisdate.isoformat(), descriptors, consensusopeners])

        resultsdf.loc[-1] = stringline.split(',')
        resultsdf.index += 1
        resultsdf.sort_index(inplace=True)

    thisdate = thisdate + datetime.timedelta(1)

```

```python
driver.close()
```

```python
resultsdf.to_csv('test.csv', index=False)
```
