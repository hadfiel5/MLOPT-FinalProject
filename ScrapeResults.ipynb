```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import datetime

import pandas as pd
import os
```

```python
year = 2025

thisdate = datetime.date(year, 3,14)

resultdsf = pd.DataFrame(columns=['game_pk', 'GameDate', 'HomeID', 'HomeRuns', 'AwayID', 'AwayRuns'])
```

```python
driver = webdriver.Chrome()

driver.implicitly_wait(1)
```

```python

while thisdate.year == year:

    print(thisdate.isoformat())

    driver.get("https://baseballsavant.mlb.com/gamefeed?date=" + thisdate.isoformat())

    driver.find_element(By.XPATH, '//div[@id="charts"]')
    games = driver.find_elements(By.XPATH, '//div[@class="game-container step"]')

    if len(games) > 0 and len(games) < 13:
        games = driver.find_elements(By.XPATH, '//div[@class="game-container step"]')

    for game in games:
        subels = game.find_elements(By.XPATH, './div')
        # Each will have three subelements, the first willhave the game pk
        gamepk = subels[0].get_attribute("name")

        status = subels[0].find_element(By.XPATH, './div[@class="game-status-container"]')
        if 'FINAL' in status.text:

            details = subels[0].find_elements(By.XPATH, './div[@class="game-information"]/div/div')

            # Away Logo is in the first
            Awaylogo = details[0].find_element(By.XPATH, './img')
            AwayID = Awaylogo.get_attribute("src").split('/')[-1]
            AwayID = int(AwayID.replace('.svg', ''))

            # Away Score is 3rd
            AwayScore = int(details[2].text)

            # Home Logo is in the last
            Homelogo = details[-1].find_element(By.XPATH, './img')
            HomeID = Homelogo.get_attribute("src").split('/')[-1]
            HomeID = int(HomeID.replace('.svg', ''))

            # Home Score is 3rd last
            HomeScore = int(details[-3].text)

            resultdsf.loc[len(resultdsf)] = [gamepk, thisdate.isoformat(), HomeID, HomeScore, AwayID, AwayScore]

    thisdate = thisdate + datetime.timedelta(1)

        
```

```python
driver.close()
```

```python
resultdsf.to_csv(os.path.join('..', 'Data', str(year), 'game_results.csv'), index=False)
```
